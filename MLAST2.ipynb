```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
dataset =pd.read_csv("/content/carcrash.csv")
dataset
```

```python
plt.figure(figsize=(20, 10))
ds=dataset.head(30)
plt.scatter(dataset.total,dataset.alcohol, label='total vs alcohol', color='blue')
plt.xlabel('total')
plt.ylabel('alcohol')
plt.title('total vs alcohol')
plt.legend()
plt.show()
```

higher the amount of alchohol , more number of incidents

```python
plt.figure(figsize=(20, 10))
ds=dataset.head(10)
plt.bar(ds.speeding,ds.alcohol, label='Bar Plot', color='green')
plt.xlabel('Speeding')
plt.ylabel('Alcohol')
plt.title('Bar Plot Example')
plt.legend()
plt.show()
```

People with high alcohol consumption end up speeding more

```python
plt.figure(figsize=(20, 10))
ds=dataset.head(30)
plt.scatter(ds.speeding,ds.not_distracted, label='Bar Plot', color='red')
plt.xlabel('Speeding')
plt.ylabel('Not distracted')
plt.title('Speeding vs Distraction')
plt.legend()
plt.show()
```

people who arent distracted tend to high a lower speeding score.

```python
plt.figure(figsize=(20, 10))
ds=dataset.head(30)
plt.pie(ds.speeding,labels=ds.not_distracted,autopct='%1.1f%%', startangle=90  )
plt.xlabel('Speeding')
plt.ylabel('Not distracted')
plt.title('Speeding vs Distraction')
plt.legend()
plt.show()
```
